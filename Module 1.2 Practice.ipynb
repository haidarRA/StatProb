# a) Range
Range adalah selisih dari nilai terbesar dan nilai terkecil dalam suatu kumpulan data. Sebelum membuat histografi, cari range dari data - data yang telah diberikan.

```python
# Fungsi untuk menghitung rentang
def calculate_range(data):
    nilai_tertinggi = max(data)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(data)
    print(f"Nilai terendah: {nilai_terendah}")

    data_range = max(data) - min(data)
    
    print(f"\n")    
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    
    return data_range


# Memasukkan data
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

data_range = calculate_range(data)
```

```python
Data: [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
Nilai tertinggi: 19
Nilai terendah: 9


Range: 19 - 9 = 10
```

Setelah mendapatkan range, kita dapat membuat histogramnya.

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

y = 1.5
x1 = 4/70*1
x2 = 66/70*1

# Membuat garis horizontal range
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(data)}')

# Membuat garis vertika range
plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')

# Menambahkan label dan judul pada histogram
plt.xlabel('Jumlah Gol')
plt.ylabel('Frekuensi')
plt.title('Range')

plt.legend()

plt.show()
```

# b) Variansi
Variansi adalah ukuran statistik yang mengukur sebaran atau variasi data dari rata-rata. Variansi dapat dihitung dengan menjumlahkan kuadrat dari selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data. Hitung variansi terlebih dahulu.

```python
# Fungsi untuk menghitung rata-rata
def calculate_mean(data):
    return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def calculate_variance(data):
    mean = calculate_mean(data)
    print(f"Rata-rata: {mean}")  
    
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")    
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean})**2 = {((x - mean)):.2f}**2 = {squared_diff_per_x:.2f}")    
    
    print(f"Total kuadrat selisih: {sum(squared_diff)}")    
    print(f"Jumlah total data: {len(data)}")    
    
    variance = sum(squared_diff) / len(data)
    
    print(f"\n")    
    print(f"Variansi: {sum(squared_diff)} / {len(data)} = {variance}")    
    
    return variance


# Memasukkan data
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

# Menghitung variansi
variance = calculate_variance(data)
```

```python
Data: [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
Rata-rata: 13.2
Kuadrat selisih setiap nilai data dengan rata-rata:
(19 - 13.2)**2 = 5.80**2 = 33.64
(10 - 13.2)**2 = -3.20**2 = 10.24
(10 - 13.2)**2 = -3.20**2 = 10.24
(12 - 13.2)**2 = -1.20**2 = 1.44
(13 - 13.2)**2 = -0.20**2 = 0.04
(15 - 13.2)**2 = 1.80**2 = 3.24
(14 - 13.2)**2 = 0.80**2 = 0.64
(13 - 13.2)**2 = -0.20**2 = 0.04
(9 - 13.2)**2 = -4.20**2 = 17.64
(17 - 13.2)**2 = 3.80**2 = 14.44
Total kuadrat selisih: 91.6
Jumlah total data: 10


Variansi: 91.6 / 10 = 9.16
```

# c) Standar Deviasi
Standar deviasi adalah ukuran statistik yang memberikan gambaran tentang seberapa jauh data tersebar dari rata-rata. Semakin tinggi standar deviasi, semakin besar variasi atau sebaran data. Standar deviasi adalah akar kuadrat dari variansi. Hitung standar deviasi terlebih dahulu.

```python
# Fungsi untuk menghitung standar deviasi
def calculate_standard_deviation(data):
    variance = calculate_variance(data)
    standard_deviation = variance**0.5
    print(f"\n")   
    print(f"Standard Deviation: {variance}**0.5 = {standard_deviation:.2f}")
    
    return standard_deviation

# Memasukkan data
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

standard_deviation = calculate_standard_deviation(data)
mean = calculate_mean(data)
```

```python
Data: [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
Rata-rata: 13.2
Kuadrat selisih setiap nilai data dengan rata-rata:
(19 - 13.2)**2 = 5.80**2 = 33.64
(10 - 13.2)**2 = -3.20**2 = 10.24
(10 - 13.2)**2 = -3.20**2 = 10.24
(12 - 13.2)**2 = -1.20**2 = 1.44
(13 - 13.2)**2 = -0.20**2 = 0.04
(15 - 13.2)**2 = 1.80**2 = 3.24
(14 - 13.2)**2 = 0.80**2 = 0.64
(13 - 13.2)**2 = -0.20**2 = 0.04
(9 - 13.2)**2 = -4.20**2 = 17.64
(17 - 13.2)**2 = 3.80**2 = 14.44
Total kuadrat selisih: 91.6
Jumlah total data: 10


Variansi: 91.6 / 10 = 9.16


Standard Deviation: 9.16**0.5 = 3.03
```

Setelah mendapatkan standar deviasi, kita dapat membuat histogramnya.

```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Membuat garis standar deviasi
plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

# Menambahkan label dan judul pada histogram
plt.xlabel('Jumlah Gol')
plt.ylabel('Frekuensi')
plt.title(f'Standar Deviasi = {standard_deviation:.2f}')

plt.legend()

plt.show()
```

Dari kedua ukuran sebaran tersebut, standar deviasi adalah ukuran sebaran yang paling cocok untuk menggambarkan sebaran data dengan baik karena pusatnya berada di garis mean dan daerahnya bergantung pada mean. Range tidak cocok karena hanya menggunakan nilai terbesar dan nilai terkecil dari seluruh data sehingga kita kurang bisa mengetahui sebaran data lebih lanjut.
