# No. 1
# a) Mean
Mean adalah rata - rata dari seluruh data yang dimasukkan. Mean dapat dihitung dengan menjumlahkan setiap data dan kemudian membaginya dengan banyak data. Sebelum membuat histogram, cari mean dari data - data yang telah diberikan.

```python
# Memasukkan data
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

# Menjumlahkan seluruh data
jumlah_nilai = sum(data)
print(f"Jumlah: {jumlah_nilai:.2f}")

# Menghitung banyaknya data
banyak_data = len(data)
print(f"Banyak: {banyak_data}")

# Menghitung mean dengan membagi jumlah_nilai dengan banyak_data
mean_value = jumlah_nilai/banyak_data
print(f"Mean: {mean_value:.2f}")
```

`
Data: [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
Jumlah: 1170.10
Banyak: 11
Mean: 106.37`

Setelah menghitung mean, kita dapat membuat histogramnya.

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=1, color='blue')

# Membuat garis mean
plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')

# Menambahkan label dan judul pada histogram
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Mean')

plt.legend()

plt.show()
```

# b) Median
Median adalah nilai tengah dari data. Median dapat dihitung dengan dua cara. Jika banyak data adalah genap, maka median dapat dihitung dengan menjumlahkan dua nilai tengah pada data, kemudian membaginya dengan 2. Sedangkan, jika banyak data adalah ganjil, maka nilai median adalah nilai tengah itu sendiri. Hitung nilai median terlebih dahulu.

```python
# Memasukkan data
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

# Mengurutkan data dari yang kecil sampai besar
sorted_data = sorted(data)
print(f"Data berurut: {sorted_data}")

# Menghitung banyaknya data
banyak_data = len(data)
print(f"Banyak: {banyak_data}")

# Mencari urutan dari nilai tengah
index_tengah = banyak_data // 2

# Jika banyak data adalah genap, maka nilai median adalah setengah dari penjumlahan dua nilai tengah
if banyak_data % 2 == 0:
    print(f"Banyaknya data adalah genap")
    median1 = sorted_data[index_tengah - 1]
    median2 = sorted_data[index_tengah]
    print(f"Nilai tengah 1: {median1}")
    print(f"Nilai tengah 2: {median2}")
    true_median = (median1 + median2) / 2
    print(f"Median: {true_median}")

# Jika banyak data adalah ganjil, maka nilai median adalah nilai tengah
else:
    print(f"Banyaknya data adalah ganjil")
    print(f"Nilai tengah: {sorted_data[index_tengah]}")
    true_median = sorted_data[index_tengah]
    print(f"Median: {true_median}")
```

`
Data: [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
Data berurut: [98.7, 98.9, 99.7, 99.9, 100, 100.5, 101, 101.3, 102.2, 102.9, 165]
Banyak: 11
Banyaknya data adalah ganjil
Nilai tengah: 100.5
Median: 100.5`

Setelah menghitung median, kita dapat membuat histogramnya.

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=1, color='blue')

# Membuat garis median
plt.axvline(x=true_median, color='red', linestyle='dashed', linewidth=1, label=f'Median = {true_median:.2f}')

# Menambahkan label dan judul pada histogram
plt.xlabel('Tinggi Badan')
plt.ylabel('Frekuensi')
plt.title('Median')

plt.legend()

plt.show()
```

# c) Modus
Modus adalah data yang paling banyak muncul dari seluruh data yang ada. Tentukan modus terlebih dahulu.

```python
from collections import Counter

# Memasukkan data
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

# Mengurutkan data dari yang kecil sampai besar
sorted_data = sorted(data)
print(f"Data berurut: {sorted_data}")

# Mencari modus
counter = Counter(sorted_data)
mode_data = dict(counter)
max_data = max(list(counter.values()))
print(f"Jumlah data modus: {max_data}")

data_modus = []
for k, v in mode_data.items():
    print(f"{k} muncul sebanyak {v} kali")
    if v == max(list(counter.values())):
        data_modus.append(k)
        
print(f"Data modus: {data_modus}")
```

Setelah mendapatkan nilai modus, kita dapat membuat histogramnya.

```python
import matplotlib.pyplot as plt

# Membuat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=1, color='blue')

# Membuat variabel penomoran modus
init = 1

# Mengurutkan data modus dari yang kecil sampai yang besar agar lebih mudah dibaca
data_modus_sorted = sorted(data_modus)

# Membuat garis modus
for i in data_modus_sorted:
    plt.axvline(x=i, color = 'red', linestyle = 'dashed', linewidth=1, alpha = 0.4, label = f"Modus {init} = {i:.2f}")
    init = init + 1

# Menambahkan label dan judul pada histogram
plt.xlabel('Data')
plt.ylabel('Frekuensi')
plt.title('Modus')

plt.legend()

plt.show()
```

Dari ketiga metode tersebut, median lebih cocok untuk menentukan pusat distribusi data. Metode mean tidak cocok digunakan karena garis mean terletak di luar kumpulan data, sedangkan metode modus tidak cocok digunakan karena masing - masing data hanya mempunyai frekuensi 1, sehingga modus dari data tinggi badan tersebut adalah semua data yang ada.

# c) Modus
Tentukan modus terlebih dahulu sebelum membuat histogram.
